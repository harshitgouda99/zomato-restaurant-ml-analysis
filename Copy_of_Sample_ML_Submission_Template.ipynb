## Project Summary

This project focuses on analyzing Zomato restaurant data to extract meaningful insights about restaurants, customer reviews, and overall dining trends. With the rapid growth of the restaurant industry and online food delivery services, analyzing such data can help businesses understand customer preferences and improve service quality. The dataset used in this project consists of two main files: the restaurant metadata dataset and the restaurant reviews dataset. The reviews dataset contains information about reviewers, review text, ratings given by customers, metadata about reviewers, time of review, and the number of pictures posted with the review. After merging the datasets, exploratory data analysis (EDA) was performed to understand patterns and trends in the data. Various visualizations were created using Matplotlib and Seaborn libraries. Restaurants were then grouped into clusters based on attributes such as cost, rating, and review characteristics. These clusters represent different segments of restaurants, such as budget restaurants, mid-range restaurants, and premium restaurants.

Finally, the clustering results were evaluated using the silhouette score to measure the quality of the clusters. The project concluded by interpreting the clusters and generating business insights. The findings from this project can help restaurant businesses understand customer preferences, improve services, and make data-driven decisions. Customers can also benefit from these insights by identifying restaurants that match their preferences in terms of cost, quality, and customer satisfaction.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Removing duplicate rows from both datasets to ensure data quality
restaurants = restaurants.drop_duplicates()
reviews = reviews.drop_duplicates()

# Checking missing values in both datasets
restaurants.isnull().sum()
reviews.isnull().sum()

# Filling missing values where appropriate
# Example: replacing missing 'Collections' values with 'Unknown'
restaurants['Collections'] = restaurants['Collections'].fillna('Unknown')

# Replacing missing 'Timings' with 'Not Available'
restaurants['Timings'] = restaurants['Timings'].fillna('Not Available')

# Renaming column 'Restaurant' in reviews dataset to match 'Name' column
reviews = reviews.rename(columns={'Restaurant': 'Name'})

# Merging restaurant metadata with review data using the restaurant name
data = pd.merge(restaurants, reviews, on='Name')

# Removing rows where review is missing
data = data.dropna(subset=['Review'])

# Now calculate review length
data['review_length'] = data['Review'].apply(len)
# Displaying the first few rows of the processed dataset
data.head()

# Checking final dataset shape after wrangling
data.shape
````

### What all manipulations have you done and insights you found?

During the data wrangling process, several steps were performed to prepare the dataset for analysis. First, both the restaurant metadata dataset and the restaurant reviews dataset were loaded and examined for their structure and data types. Duplicate records were identified and removed to ensure that the dataset contained only unique entries. Missing values were analyzed using null value checks and visualized to understand which variables had incomplete information.

The two datasets were then merged using the restaurant name as the common column so that restaurant details and customer reviews could be analyzed together. After merging, additional features such as review length were created to extract more meaningful information from the review text. These transformations helped organize the data and make it suitable for further exploratory data analysis and machine learning tasks.

From this process, it was observed that each restaurant can have multiple reviews, which provides valuable information about customer opinions. #### Chart - 3

````python
#Bivariate Analysis
# Function to calculate sentiment polarity from review text
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

# Creating sentiment score column
data['sentiment_score'] = data['Review'].apply(get_sentiment)

# Converting sentiment score into categories
def sentiment_label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

# Creating sentiment category column
data['sentiment'] = data['sentiment_score'].apply(sentiment_label)
# Chart - 3 : Relationship between Sentiment and Rating

plt.figure(figsize=(8,5))

# Boxplot to compare rating distribution across sentiment categories
sns.boxplot(x='sentiment', y='Rating', data=data)

plt.title("Sentiment vs Restaurant Rating")
plt.xlabel("Sentiment Category")
plt.ylabel("Rating")

plt.show()
````

##### 1. Why did you pick the specific chart?

A boxplot is useful for comparing the distribution of a numerical variable across different categories. In this case, it helps analyze how restaurant ratings vary across different sentiment categories such as positive, negative, and neutral reviews. This visualization makes it easier to understand the relationship between customer sentiment and the ratings they give.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that positive sentiment reviews generally correspond to higher ratings, while negative sentiment reviews are associated with lower ratings. Neutral reviews tend to fall around the middle rating range. This indicates a strong relationship between the sentiment expressed in customer reviews and the ratings given.

````python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3,5,10,None],
    'min_samples_split': [2,5,10]
}

grid_dt = GridSearchCV(DecisionTreeRegressor(random_state=42),
                       param_grid,
                       cv=3,
                       scoring='r2')

grid_dt.fit(X_train, y_train)

best_dt = grid_dt.best_estimator_

y_pred_dt = best_dt.predict(X_test)

print("Best Parameters:", grid_dt.best_params_)
print("Updated R2 Score:", r2_score(y_test, y_pred_dt))
````
